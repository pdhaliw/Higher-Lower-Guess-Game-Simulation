# Number Guessing Strategy Simulation

This notebook analyzes multiple strategies for solving the number guessing game:

A number between 1 and N is selected.  
After each guess, feedback is given:
- "Higher" if the guess is too low  
- "Lower" if the guess is too high  

The process continues until the correct number is guessed.

---

## Method Descriptions

### Linear Search
Always guesses the lowest remaining number.  
Worst-case complexity: O(N).  
Deterministic strategy.

### Reverse Linear Search
Always guesses the highest remaining number.  
Worst-case complexity: O(N).  
Deterministic strategy.

### Random
Selects a random number within the current bounds.  
Expected complexity: O(N).  
High variance across runs.

### Binary Search
Always guesses the midpoint of the interval.  
Worst-case complexity: O(log₂N).  
Optimal deterministic strategy for uniformly distributed targets.

### Ternary Search
Splits the interval into thirds.  
Typically less efficient than binary search for this problem.

### Golden Ratio Search
Partitions the interval using an approximately 0.618 proportion.  
Inspired by numerical optimization methods.

### Noisy Midpoint
Uses the midpoint plus small random variation.  
Models imperfect human guessing.

### Step 25
Guesses in increments of 25 before refining the interval.  
Represents a simple stepping heuristic.

---

## Notebook Structure

This notebook supports two experimental modes.

### 1. Individual Method Runs

Each strategy can be executed independently using:

    run_method(...)

This produces a dataframe containing:
- method
- target_number
- guesses

These individual dataframes can later be combined using:

    pd.concat([...])

This approach is useful for evaluating standalone performance.

---

### 2. Shared Target Multi-Method Runs

All methods can be executed together using:

    run_all_methods_shared_target(...)

In this mode:
- Each run generates one random target number.
- All methods attempt that same number.
- Output includes:
    - run_number
    - target_number
    - method
    - guesses

This allows controlled comparison across strategies.

# Main Engine

```python
def simulate_game(guesser_func, lower=1, upper=100):
    target = random.randint(lower, upper)
    low = lower
    high = upper
    guesses = 0

    while True:
        guess = guesser_func(low, high)
        guesses += 1

        if guess == target:
            return target, guesses
        elif guess < target:
            low = guess + 1
        else:
            high = guess - 1


def run_method(guesser_func, method_name, n_runs=1000, lower=1, upper=100):
    results = []
    for _ in range(n_runs):
        target, guesses = simulate_game(guesser_func, lower, upper)
        results.append({
            "method": method_name,
            "target_number": target,
            "guesses": guesses
        })
    return pd.DataFrame(results)
```

## TERNARY SEARCH SECTION

```python
def ternary_guesser(low, high):
    return low + (high - low) // 3
```

```python
ternary_results = run_method(ternary_guesser, "Ternary Search", n_runs=1000)
```

-----------------

## GOLDEN RATIO SEARCH SECTION

```python
def golden_ratio_guesser(low, high):
    phi = 0.618
    return low + int(phi * (high - low))
```

```python
golden_results = run_method(golden_ratio_guesser, "Golden Ratio", n_runs=1000)
```

```python
golden_results.head()
```

-------------

## NOISY MIDPOINT SECTION

```python
def noisy_midpoint_guesser(low, high):
    mid = (low + high) // 2
    noise = random.randint(-3, 3)
    guess = max(low, min(high, mid + noise))
    return guess
```

```python
noisy_results = run_method(noisy_midpoint_guesser, "Noisy Midpoint", n_runs=1000)
```

```python
noisy_results.head()
```

-------------

## STEP-25 SECTION

```python
def step_25_guesser(low, high):
    step = 25
    guess = low + step
    if guess > high:
        guess = (low + high) // 2
    return guess
```

```python
step_results = run_method(step_25_guesser, "Step 25", n_runs=1000)
```

```python
step_results.head()
```

### This is for a targeted multi-method runner

```python
def simulate_fixed_target(guesser_func, target, lower=1, upper=100):
    low = lower
    high = upper
    guesses = 0

    while True:
        guess = guesser_func(low, high)
        guesses += 1

        if guess == target:
            return guesses
        elif guess < target:
            low = guess + 1
        else:
            high = guess - 1
```

```python
def run_all_methods_shared_target(n_runs=1000, lower=1, upper=100):

    guessers = {
        "Linear": linear_guesser,
        "Reverse Linear": reverse_linear_guesser,
        "Random": random_guesser,
        "Binary": binary_guesser,
        "Ternary": ternary_guesser,
        "Golden Ratio": golden_ratio_guesser,
        "Noisy Midpoint": noisy_midpoint_guesser,
        "Step 25": step_25_guesser
    }

    results = []

    for run in range(1, n_runs + 1):

        # NEW RANDOM NUMBER EACH RUN
        target = random.randint(lower, upper)

        for method_name, guesser_func in guessers.items():

            guesses = simulate_fixed_target(
                guesser_func,
                target,
                lower,
                upper
            )

            results.append({
                "run_number": run,
                "target_number": target,
                "method": method_name,
                "guesses": guesses
            })

    return pd.DataFrame(results)
```

```python
shared_results = run_all_methods_shared_target(n_runs=100000)
shared_results.head()
```

```python
# Total rows
total_rows = shared_results.shape[0]

# Number of unique methods
num_methods = shared_results["method"].nunique()

# Compute number of runs
n_runs = total_rows // num_methods

print("Total rows:", total_rows)
print("Methods:", num_methods)
print("Computed runs:", n_runs)
```

# Purpose of the Experiment

This simulation framework allows structured comparison of search strategies under identical rules.

Two testing modes are provided:

## Independent Testing
Each method runs on its own randomly generated targets.
Useful for measuring overall average behavior.

## Shared Target Testing
Each run uses a single randomly generated number shared across all methods.
This removes variation in target selection between methods and enables direct comparison.

# Theoretical Insight

Binary search minimizes worst-case guesses for uniformly distributed targets and achieves logarithmic complexity.

Other strategies demonstrate:
- Linear inefficiency
- High variance behavior
- Heuristic-based interval reduction
- Alternative partitioning logic

This notebook serves as a benchmarking framework for evaluating search algorithms in a controlled environment.
